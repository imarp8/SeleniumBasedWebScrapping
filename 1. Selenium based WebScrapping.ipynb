```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import pandas as pd
import time

# -------------------------------
# 🔒 Setup Chrome
# -------------------------------
chrome_options = Options()
chrome_options.add_argument("--disable-notifications")
chrome_options.add_experimental_option(
    "prefs", {"profile.default_content_setting_values.notifications": 2}
)

driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()),
    options=chrome_options
)

wait = WebDriverWait(driver, 15)

# -------------------------------
# 🔍 Input
# -------------------------------
ean = "3600523966905"

# -------------------------------
# 🌐 Open Website
# -------------------------------
driver.get("https://www.superdrug.com")

# Accept cookies (if present)
try:
    wait.until(EC.element_to_be_clickable(
        (By.XPATH, "//button[contains(text(),'Accept')]"))
    ).click()
except:
    pass

# -------------------------------
# 🔍 Perform Search
# -------------------------------
search_box = wait.until(
    EC.presence_of_element_located(
        (By.XPATH, "//input[@role='combobox' and @aria-label='Search']")
    )
)

search_box.clear()
search_box.send_keys(ean)
search_box.send_keys(Keys.RETURN)

# Wait for results
wait.until(EC.presence_of_element_located(
    (By.XPATH, "//a[contains(@class,'cx-product-name')]"))
)

# Remove chat popup if present
driver.execute_script("""
var elem = document.getElementById('toky_container');
if (elem) { elem.remove(); }
""")

# Scroll to trigger lazy loading
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
time.sleep(3)

# -------------------------------
# 🧠 Extract data (GLOBAL LIST METHOD)
# -------------------------------

# Product names + URLs
names = driver.find_elements(By.XPATH, "//a[contains(@class,'cx-product-name')]")

# Promotion label (your first element)
labels = driver.find_elements(By.XPATH, "//span[contains(@class,'promotion__offer-label')]")

# Promotion detail (your second element)
details = driver.find_elements(By.XPATH, "//div[contains(@class,'promotion__label')]")

print("DEBUG COUNTS:")
print("Products:", len(names))
print("Labels:", len(labels))
print("Details:", len(details))

# -------------------------------
# 🔁 Map data using index
# -------------------------------
results = []

max_len = max(len(names), len(labels), len(details))

for i in range(max_len):

    # Product
    name = names[i].text.strip() if i < len(names) else "No name"
    url = names[i].get_attribute("href") if i < len(names) else "No URL"

    # Promotion
    label = labels[i].text.strip() if i < len(labels) else "No offer"
    detail = details[i].text.strip() if i < len(details) else "No detail"

    results.append({
        "EAN": ean,
        "Product Name": name,
        "URL": url,
        "Offer Label": label,
        "Offer Detail": detail
    })

# -------------------------------
# 📊 Convert to pandas
# -------------------------------
df = pd.DataFrame(results)

print("\n✅ Final Output:")
display(df.head())

driver.quit()
```

```python
# Using global label indexing is leaking promotion codes to other product ids.

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import pandas as pd
import time

# -------------------------------
# 🔒 Setup Chrome
# -------------------------------
chrome_options = Options()
chrome_options.add_argument("--disable-notifications")
chrome_options.add_experimental_option(
    "prefs", {"profile.default_content_setting_values.notifications": 2}
)

driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()),
    options=chrome_options
)

wait = WebDriverWait(driver, 15)

# -------------------------------
# 🔍 Input
# -------------------------------
ean = "3600524049102"

# -------------------------------
# 🌐 Open Website
# -------------------------------
driver.get("https://www.superdrug.com")

# Accept cookies
try:
    wait.until(EC.element_to_be_clickable(
        (By.XPATH, "//button[contains(text(),'Accept')]"))
    ).click()
except:
    pass

# -------------------------------
# 🔍 Perform Search
# -------------------------------
search_box = wait.until(
    EC.presence_of_element_located(
        (By.XPATH, "//input[@role='combobox' and @aria-label='Search']")
    )
)

search_box.clear()
search_box.send_keys(ean)
search_box.send_keys(Keys.RETURN)

# Wait for product cards
wait.until(EC.presence_of_element_located(
    (By.XPATH, "//div[contains(@class,'product-list-item')]"))
)

# Remove chat popup
driver.execute_script("""
var elem = document.getElementById('toky_container');
if (elem) { elem.remove(); }
""")

# Scroll to trigger lazy loading
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
time.sleep(3)

# -------------------------------
# 🧠 Extract products + promotions
# -------------------------------
products = driver.find_elements(
    By.XPATH, "//div[contains(@class,'product-list-item')]"
)

results = []

for p in products:

    # -------------------------------
    # ✅ Validate product (avoid ghost rows)
    # -------------------------------
    name_elements = p.find_elements(
        By.XPATH, ".//a[contains(@class,'cx-product-name')]"
    )

    if not name_elements:
        continue

    name = name_elements[0].get_attribute("innerText").strip()

    if not name:
        continue

    url = name_elements[0].get_attribute("href")

    if not url:
        continue

    # -------------------------------
    # 🔥 Extract promotions (multiple)
    # -------------------------------
    promo_blocks = p.find_elements(By.XPATH, ".//e2-promotion")

    valid_promo_found = False

    for block in promo_blocks:

        try:
            label = block.find_element(
                By.XPATH, ".//span[contains(@class,'promotion__offer-label')]"
            ).get_attribute("innerText").strip()
        except:
            label = ""

        try:
            detail = block.find_element(
                By.XPATH, ".//div[contains(@class,'promotion__label')]"
            ).get_attribute("innerText").strip()
        except:
            detail = ""

        # ❌ Skip empty promotions
        if not label and not detail:
            continue

        valid_promo_found = True

        results.append({
            "EAN": ean,
            "Product Name": name,
            "URL": url,
            "Offer Label": label,
            "Offer Detail": detail
        })

    # ❌ Skip products with no valid promotions
    if not valid_promo_found:
        continue

# -------------------------------
# 📊 Convert to pandas
# -------------------------------
df = pd.DataFrame(results)

# Remove duplicates (extra safety)
df = df.drop_duplicates(subset=["Product Name", "URL", "Offer Label", "Offer Detail"])

print("\n✅ FINAL CLEAN OUTPUT:")
print(df.head())

driver.quit()
```
